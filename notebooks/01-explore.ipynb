## Q4: How many distinct values does Fatal (Y/N) have? What's wrong with it?

```python
df['Fatal (Y/N)'].unique()
```

```python
from numpy import nan


df['Fatal (Y/N)']= df['Fatal (Y/N)'].replace([' N','N ','n'],'N')

df['Fatal (Y/N)']= df['Fatal (Y/N)'].replace('F','Y')

df['Fatal (Y/N)']= df['Fatal (Y/N)'].replace([nan,'2017','#VALUE!'],'UNKNOWN')
```
