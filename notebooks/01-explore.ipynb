-  Q3: What % of Age values are missing? What % of Time?

```python
null_pct = (df.isnull().sum() / len(df) * 100).round(2).sort_values(ascending=False)
null_pct
```

```python
null_pct = (df_removed_nulls.isnull().sum() / len(df_removed_nulls) * 100).round(2).sort_values(ascending=False)
null_pct
```
